```python
# Make sure this comes before the matplotlib import
%matplotlib inline

import matplotlib.pyplot as plt
# Same As
from matplotlib import pyplot as plt
```

```python
# Plot y = x ^ 2 -x + 2
x = range(-5, 6)
y = []
for n in x:
    y.append(n ** 2 - n + 2)

print(x)
print(y)
```

```python
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = x^2 - x + 2')
plt.plot(x, y, ls=':')
plt.text(0, 4, '(0, 0)', fontsize=10)
```

```python
from math import sqrt, tan

x1 = range(0, 8)
x2 = range(-7, 8)
x3 = x1
x4 = x2

# y1 = sqrt x
y1 = []
for x in x1:
    y1.append(sqrt(x))
# y2 = x ^ 2
y2 = []
for x in x2:
    y2.append(x ** 3)
# y3 = tan x
y3 = []
for x in x3:
    y3.append(tan(x))
# y4 = 2 ^ x
y4 = []
for x in x4:
    y4.append(2 ** x)
```

```python
plt.figure(figsize=(12, 8)) # (width, height)

plt.suptitle('Let\'s see if I can remember algebra from many years ago')

plt.subplot(221)
plt.plot(x1, y1)
plt.title('y = sqrt(x)')

plt.subplot(222)
plt.plot(x2, y2)
plt.title('y = x ^ 3')

plt.subplot(223)
plt.plot(x3, y3)
plt.title('y = tan(x)')

plt.subplot(224)
plt.plot(x4, y4)
plt.title('y = 2^x')
```

```python
plt.figure(figsize=(12, 8)) # (width, height)

plt.plot(x1, y1, label='y = sqrt(x)')
plt.plot(x2, y2, label='y = x^3')
plt.plot(x3, y3, label='y = tan(x)')
plt.plot(x4, y4, label='y = 2^x')

plt.legend()

plt.title('various x and y values')
```
