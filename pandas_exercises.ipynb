```python
# for long lines:
# - store each intermediate calculation in a variable

by_manufacturer = cars[['manufacturer', 'hwy', 'cty']].groupby('manufacturer')
avg_by_manufacturer = by_manufacturer.mean()
avg_by_manufacturer.sort_values(by='hwy', ascending=False)

# - backslashes to continue the expression over many lines
cars[['manufacturer', 'hwy', 'cty']]\
    .groupby('manufacturer')\
    .mean()\
    .sort_values(by='hwy', ascending=False)

# - parenthesis
(cars[['manufacturer', 'hwy', 'cty']]
 .groupby('manufacturer')
 .mean()
 .sort_values(by='hwy', ascending=False))

cars.assign(better_avg_mileage=cars.cty * 0.55 + cars.hwy * 0.45,
            avg_mileage=(cars.cty + cars.hwy) / 2)\
    [['manufacturer', 'avg_mileage', 'better_avg_mileage']]\
    .groupby('manufacturer')\
    .mean()\
    .sort_values('avg_mileage', ascending=False)
```

- What is the overal percentage of specials?
- How many animals are hoppers that are above the median speed? What percentage is this?

```python
mammals.shape
```

```python
# what is the weight of the fastest animal?

mammals.sort_values(by='speed', ascending=False).head(1)

mammals.loc[mammals.speed.idxmax()]['weight']

mammals.speed.max()
mammals.speed.idxmax()
mammals.loc[53]['weight']

mammals[mammals.speed == mammals.speed.max()]

mammals.nlargest(1, 'speed')
```

```python
n_specials = mammals[mammals.specials].shape[0]
n_total = mammals.shape[0]

n_specials / n_total

mammals.specials.sum() / mammals.weight.count()

percent_specials = mammals.specials.sum() / len(mammals)

print('{:.2%} are specials'.format(percent_specials))
```

```python
# hoppers over the median speed

hoppers = mammals[mammals.hoppers]
median_speed = mammals.speed.median()

n_hoppers = hoppers[hoppers.speed > median_speed].shape[0]

percent_hoppers = n_hoppers / mammals.shape[0]

print('There are {} hoppers, which is {:.2%} of the total'.format(n_hoppers, percent_hoppers))
```
